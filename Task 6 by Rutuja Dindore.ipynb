## Importing required Libraries 

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set()
import warnings 
warnings.filterwarnings('ignore')
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('Iris.csv',index_col=0)
```

Undestanding the data 

```python
df.head()
```

```python
df.isnull().sum().any()
```

```python
df.shape
```

```python
df.describe()
```

```python
#Checking the unique values in target variable 
df.Species.unique()
```

```python
#splitting our data for training and testing 
x=df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].values
```

```python
y=df.Species.values
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
#creating the model 
model=tree.DecisionTreeClassifier(criterion='entropy')
```

```python
#training our classifier 
model.fit(x_train,y_train)
```

```python
#predicting using the  classifier
y_pred= model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

### Evaluation 

```python
from sklearn.metrics import accuracy_score,classification_report
accuracy_score(y_test,y_pred)
```

```python
#Evaluation Summary 
print(classification_report(y_test,y_pred))
```

## Decision Tree Visualization

```python
cn = df["Species"].unique().tolist()
fig = plt.figure(figsize=(25,20))
tree.plot_tree(model, feature_names=df.columns[:-1],class_names=cn,filled=True)
```
